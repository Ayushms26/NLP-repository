```python
exclude = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
```

```python
def remove_punc(text):
    for i in exclude:
        text = text.replace(i,'') 
        # if i vala koi punctuation text me hoga then replace it by ''.
    return text
```

```python
import time
start = time.time()
text = 'My name &* is\\! ay*us@@h}}'

remove_punc(text)

print((time.time() - start)*50000) # if apply on our 50k row data
```

```python
# but working with large dataset, that function remove_punc get too show
# hence there is one more way

def remove_punc1(text):
    return text.translate(str.maketrans('','',exclude))
```

```python
chat_words = dict{}
```

```python
chat_words = {'AFAIK' : 'As Far As I Know',
'AFK' : 'Away From Keyboard',
'ASAP' : 'As Soon As Possible',
'ATK' : 'At The Keyboard',
'ATM' : 'At The Moment',
'A3' : 'Anytime, Anywhere, Anyplace',
'BAK' : 'Back At Keyboard',
'BBL' : 'Be Back Later',
'BBS' : 'Be Back Soon',
'BFN' : 'Bye For Now',
'B4N' : 'Bye For Now',
'BRB' : 'Be Right Back',
'BRT' : 'Be Right There',
'BTW' : 'By The Way',
'B4' : 'Before',
'B4N' : 'Bye For Now',
'CU' : 'See You',
'CUL8R' : 'See You Later',
'CYA' : 'See You',
'FAQ' : 'Frequently Asked Questions',
'FC' : 'Fingers Crossed',
'FWIW' : 'For What Its Worth',
'FYI' : 'For Your Information',
'GAL' : 'Get A Life',
'GG' : 'Good Game',
'GN' : 'Good Night',
'THX' : 'Thank You',
'U' : 'You',
'U2' : 'You Too'}
```

```python
def chat_conversion(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat_words:
            new_text.append(chat_words[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

```python
# Remove stop words (a, you, i , me, the, of etc)

from nltk.corpus import stopwords
```

```python
stopwords = stopwords.words('english')
```

```python
# code similer as removing punctuation

def remove_stopwords(text):
    new_text = []
    for word in text.split():
        if word in stopwords:
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
remove_stopwords(df['review'][0])
```

```python
df['review'].apply(remove_stopwords)
```

```python
# remove emoji

def remove_emoji(text):
    emoji_pattern = re.compile("["
                              u"\U0001F600-\U0001F64F"
                               u"\U0001F300-\U0001F5FF"
                               u"\U0001F680-\U0001F6FF"
                               u"\U0001F1E0-\U0001F1FF"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'',text)
                              
```

```python
# Stemming  (Do, Doing, Done -----> Do)
# For English stemming of english ----> use porterStemmer

from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

```python
sample = 'walk walks walking walked'
stem_words(sample)

# problem
# the problem is that ki -----> stemming karne ke bad jo word banta hai
# it is not nesseceary that vo uss language me exist karta ho

# EX:   stem(movie) = movi
```

```python
# A better way of stemming
# Lemmatization (slow because it find valid word in language )
# use lemmatization when tumhe output user ko show karna hai


```

```python
import nltk
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
sent = 'He was running very fast also eating fast and reading very slow. after swimming he play.'

punctuation ="?:!.,;"
sent_words = nltk.word_tokenize(sent)
for word in sent_words:
    if word in punctuation:
        sent_words.remove(word)

sent_words
```

```python
print("{0:20}{1:20}".format("Word","Lemma"))
for word in sent_words:
    print("{0:20}{1:20}".format(word, wordnet_lemmatizer.lemmatize(word,pos='v')))
```
