```python
import gensim
import os
import nltk
nltk.download('punkt')
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
import nltk

# Download the required resource
nltk.download('punkt_tab')

story = []
with open('/content/001ssb.txt', 'r', encoding='utf-8') as f:
    corpus = f.read()

raw_sent = sent_tokenize(corpus)  # Now works without errors

for sent in raw_sent:
    story.append(simple_preprocess(sent))
```

```python
story # all words with tokenize also with simple preprocessing
```

```python
model = gensim.models.Word2Vec(
    window=10, # target ki dono side there are 10-10 context word
    min_count=2, # take only those sentence jinme minimum 2 words hai
    vector_size=50
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.wv.most_similar('daenerys')
```

```python
model.wv.doesnt_match(['youth','mya','eunuch'])
```

```python
model.wv['jon'] # find vector representation for a perticuler vector
```

```python
model.wv.get_normed_vectors() # get all vectors --> shape = (7432, 50)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
```

```python
x = pca.fit_transform(model.wv.get_normed_vectors())
```

```python
x.shape
```

```python
import plotly.express as px

# Assuming x is a DataFrame or array with at least 3 columns
fig = px.scatter_3d(
    x[:150],  # Use first 100 rows
    x=0,      # Column 0 for x-axis
    y=1,      # Column 1 for y-axis
    z=2,      # Column 2 for z-axis
    color=y[:150]  # Match length of x[:100] (100 elements)
)
fig.show()
```
